```python
import math
def cosSimilar(v1,v2):

    sum = 0
    lv1 = 0
    lv2 = 0
    for i in range(v1.shape[0]):
        sum += v1[i]*v2[i]
        lv1 += v1[i]*v1[i]
        lv2 += v2[i]*v2[i]
    dp = sum/(math.sqrt(lv1)*math.sqrt(lv2))
    return dp
```

```python
def cosSimilar2(A,B):
    res = np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))
    return res
```

```python
def euclidianSimilarity(v1,v2):
    sum = 0
    for i in range(v1.shape[0]):
        sum += (v1[i]-v2[i])*(v1[i]-v2[i])
    res = 1/(1+math.sqrt(sum))
    #res = math.sqrt(sum)
    return res
```

# **Lab Exercise 4: Basic Vectorization**

In this lab exercise, we will vectorize the news dataset corpus using 
one-hot encoding and bag of words.

1.   For the newsdataset created from Lab 2:

*   Apply preprocessing.
*   Create one-hot encoded vectors for each token in the vocabulary.



2.   Consider each news article in the dataset as a document.

*   Apply preprocessing.
*   Create BoWs vectors for each of the documents



3.   Read a search text from the user

*   Using cosine similarity: List the top five similar documents based on the 
search text


```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the news dataset corpus
nltk.download("brown")  # Download the Brown corpus
news_dataset = nltk.corpus.brown.sents(categories=["news"])
news_dataset = [" ".join(document) for document in news_dataset]

# Step 1: Preprocessing and One-Hot Encoding

# Apply preprocessing to the news dataset
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

preprocessed_news_dataset = []
for document in news_dataset:
    # Tokenize the document into words
    words = word_tokenize(document)
    # Remove stopwords and lemmatize words
    words = [lemmatizer.lemmatize(word.lower()) for word in words if word.isalnum() and word.lower() not in stop_words]
    # Join the words back into a single string
    preprocessed_news_dataset.append(" ".join(words))

# Create a CountVectorizer for one-hot encoding
one_hot_vectorizer = CountVectorizer(binary=True)
one_hot_vectors = one_hot_vectorizer.fit_transform(preprocessed_news_dataset)

# Step 2: Preprocessing and Bag of Words (BoWs)

# Apply preprocessing to each news article in the dataset
preprocessed_documents = []
for document in news_dataset:
    # Tokenize the document into words
    words = word_tokenize(document)
    # Remove stopwords and lemmatize words
    words = [lemmatizer.lemmatize(word.lower()) for word in words if word.isalnum() and word.lower() not in stop_words]
    # Join the words back into a single string
    preprocessed_documents.append(" ".join(words))

# Create a CountVectorizer for BoWs
bow_vectorizer = CountVectorizer()
bow_vectors = bow_vectorizer.fit_transform(preprocessed_documents)

# Step 3: Read search text and calculate cosine similarity

# Read search text from the user
search_text = input("Enter search text: ")

# Preprocess the search text
search_text_words = word_tokenize(search_text)
search_text_words = [lemmatizer.lemmatize(word.lower()) for word in search_text_words if word.isalnum() and word.lower() not in stop_words]
search_text = " ".join(search_text_words)

# Vectorize the search text using the same vectorizer used for one-hot encoding or BoWs
if one_hot_vectorizer.vocabulary_.get(search_text) is not None:
    search_vector = one_hot_vectorizer.transform([search_text])
else:
    search_vector = bow_vectorizer.transform([search_text])

# Calculate cosine similarity between search vector and document vectors
similarity_scores = cosine_similarity(search_vector, bow_vectors)

# Get indices of top 5 similar documents
top_indices = similarity_scores.argsort()[0][-5:][::-1]

# Print the top 5 similar documents
print("Top 5 similar documents based on the search text:")
for index in top_indices:
    print(news_dataset[index])
```
